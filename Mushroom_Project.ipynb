# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
x = df.drop(["class"], axis=1)
y_new = df["class"].to_numpy()
y_new.shape
y_new = y_new.reshape(-1,1)
y = pd.DataFrame(y_new)
y
```

```python
from sklearn.preprocessing import power_transform,StandardScaler
x=power_transform(x,method='yeo-johnson')
x
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = .20, random_state = 40)
```

In the above case, we have used 80% of the data for Training and 20% of the data for Testing.

```python
train_x.shape
```

In the above case, Rows = 5254 and Column = 21.

```python
train_y.shape
```

in the above case, Rows = 5254 and Column = 1

```python
lr = LogisticRegression()
lr.fit(train_x, train_y)
```

in the above case, we are using Logistic Regression for both the Training of x and y.

```python
pred_x = lr.predict(test_x)
pred_x
```

```python
print(pred_x)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Accuracy Score", accuracy_score(test_y,pred_x))
```

```python
print("Confusion Matrix", confusion_matrix(test_y,pred_x))
```

```python
print("Classification Report", classification_report(test_y,pred_x))
```

From the above case, the Accuracy is 96% using the Logistic Regression.

```python
x_axis_labels = ["Edible Type", "Poisonous Type"]
Y_axis_labels = ["Edible Type", "Poisonous Type"]
plt.figure(figsize=(8,4))
sns.heatmap(confusion_matrix(test_y,pred_x), annot = True)
plt.show()
```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(train_x, train_y)
dt.score(train_x, train_y)
pred_dt = dt.predict(test_x)
print("Accuracy Score", accuracy_score(test_y, pred_dt))
```

```python
print("Confusion Matrix", confusion_matrix(test_y, pred_dt))
```

```python
print("Classification Report", classification_report(test_y, pred_dt))
```

From the above case, the Accuracy is 100%

```python
x_axis_labels = ["Edible Type", "Poisonous Type"]
Y_axis_labels = ["Edible Type", "Poisonous Type"]
plt.figure(figsize=(8,4))
sns.heatmap(confusion_matrix(test_y,pred_dt), annot = True)
plt.show()
```

# K-Neibors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn_class = KNeighborsClassifier(n_neighbors = 6)
kn_class.fit(train_x, train_y)
kn_class.score(train_x, train_y)
pre_kn_class = kn_class.predict(test_x)
print("Accuracy Score", accuracy_score(test_y, pre_kn_class))
```

```python
print("Confusion Matrix", confusion_matrix(test_y, pre_kn_class))
```

```python
print("Classification Report", classification_report(test_y, pre_kn_class))
```

From the above case, the Accuracy is 100%
